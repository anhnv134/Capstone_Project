##**Flow Chart Creation**

This python script is designed to create flow chart from Algorithm 1 to Algorithm 9.

```python
!pip install graphviz
```

```python
#Flow Chart Creation

# import necessary libraries
from graphviz import Digraph

def create_flowchart():
    dot = Digraph(comment='Intraday Breakout Strategy Flowchart', format='png')

    # here we define nodes for each algorithm step
    dot.node('A', 'Algorithm 1: Data Collection & Filtering\n(Liquidity, ATR, ADX)')
    dot.node('B', 'Algorithm 2: Fibonacci Levels\n(Morning Swing Computation)')
    dot.node('C', 'Algorithm 3: SI & ASI Calculation\n(Detect HSP/LSP)')
    dot.node('D', 'Algorithm 4: Breakout Confirmation & Entry\n(MACD, RSI, EMA, Volume)')
    dot.node('E', 'Algorithm 5: Dynamic Position Management\n(Trailing Stops, Partial Exits)')
    dot.node('F', 'Algorithm 6: Performance Tracking\n(Post-Session Metrics)')
    dot.node('G', 'Algorithm 7: Backtesting\n(In-Sample Evaluation)')
    dot.node('H', 'Algorithm 8: Parameter Optimization\n(Grid Search)')
    dot.node('I', 'Algorithm 9: Parameter Optimization\n(Hyperopt), Forward Testing & Paper Trading\n(Out-of-Sample & Live Simualtion)')

    # connect the nodes with descriptive lables,
    dot.edge('A', 'B', label='Clean & Merge Data')
    dot.edge('B', 'C', label='Compute Fib Levels')
    dot.edge('C', 'D', label='Confirm Breakout Conditions')
    dot.edge('D', 'E', label='Enter Trade & Manage')
    dot.edge('E', 'F', label='Record & Analyze Trades')
    dot.edge('F', 'G', label='Backtest on Historical Data')
    dot.edge('G', 'H', label='Optimize Parameters')
    dot.edge('H', 'I', label='Apply Best Params to OOS & Live Trading')

    # render and display the flowchart
    dot.render('intraday_breakout_flowchart', view=True)
    return dot

# generate the flowchart
flowchart = create_flowchart()
flowchart


```
